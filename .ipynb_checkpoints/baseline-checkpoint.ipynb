## 3. Preprocessing

### 기본정보와 QdE, QdA만 남기고 변수 제거

```python
drop_val = ['QaA', 'QbA', 'QbE', 'QcA', 'QcE', 'QeA', 'QeE',
       'QfA', 'QfE', 'QgA', 'QgE', 'QhA', 'QhE', 'QiA', 'QiE', 'QjA', 'QjE',
       'QkA', 'QkE', 'QlA', 'QlE', 'QmA', 'QmE', 'QnA', 'QnE', 'QoA', 'QoE',
       'QpA', 'QpE', 'QqA', 'QqE', 'QrA', 'QrE', 'QsA', 'QsE', 'QtA', 'QtE',
       'tp01', 'tp02', 'tp03', 'tp04', 'tp05',
       'tp06', 'tp07', 'tp08', 'tp09', 'tp10', 'urban', 'wf_01',
       'wf_02', 'wf_03', 'wr_01', 'wr_02', 'wr_03', 'wr_04', 'wr_05', 'wr_06',
       'wr_07', 'wr_08', 'wr_09', 'wr_10', 'wr_11', 'wr_12', 'wr_13']

train = train.drop(drop_val, axis = 1)
test = test.drop(drop_val, axis = 1)
train.head()
test.head()
```

### 결측치 처리

```python
print('train : ', train.isnull().sum())
print('test : ', test.isnull().sum())
```

### Label Encoding

```python
label = LabelEncoder()
drop_val2 = ['gender', 'age_group', 'race', 'religion']
for D in [train, test]:
    D['gender_code'] = label.fit_transform(D['gender'])
    D['age_group_code'] = label.fit_transform(D['age_group'])
    D['race_code'] = label.fit_transform(D['race'])
    D['religion_code'] = label.fit_transform(D['religion'])
    
train = train.drop(drop_val2, axis=1)
test = test.drop(drop_val2, axis=1)
```

#### 가설2 : 특정 성향의 사람들은 답변 응답시간이 더 길 것이다.

```python
def plot_to_voted(data, var, target):
    plt.figure(figsize=(8,6))
    for tar in target:
        u = np.percentile(data[data['voted']==tar][var], 95)
        d = np.percentile(data[data['voted']==tar][var], 5)
        column = [val for val in data[data['voted']==tar][var] if d < val < u]
        sns.distplot(column)
    plt.title(var)
    plt.legend(target)
    plt.show()
```

```python
plot_to_voted(train, 'QaE', train_y.unique())
```
